# ðŸ§¹ Data Cleaning

```python
#new columns

new_columns = ["time_stamp","gender","age","course","current_year","cgpa","married","depression","anxiety","panic_attack","specialist_treatment"]
```

```python
#Set the new columns

df.columns = new_columns
```

```python
#view the changes in the dataframe

df
```

```python
#Function to edit the time stamp

def clean(time_stamp):
   return time_stamp.split(" ")[0]
```

```python
#Apply the function

df.time_stamp = df.time_stamp.apply(clean)
```

```python
#Convert the datatype

df.time_stamp = pd.to_datetime(df.time_stamp)
```

```python
df.info()
```

```python
#Edit the time column to have the same format

def clean(date):
    return date.strftime('%d-%B-%Y').split("-")[1]
```

```python
#create the month column

df['month_filled'] = df.time_stamp.apply(clean)
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
#type cast the age column to integer

df['age'] = df.age.astype('int')
```

```python
#Edit the courses column

def clean_column(word):
    return word.strip().lower()
```

```python
df['course'] = df['course'].apply(clean_column)
```

```python
#function to edit the course column

def clean_course(course):
    if course == 'engin' or course == 'engine':
        return "engineering"
    elif course == "mathemathics":
        return "mathematics"
    else:
        return course
```

```python
df['course'] = df['course'].apply(clean_course)
```

```python
df.course.value_counts().keys()
```

```python
df.course.value_counts()
```

```python
df['current_year'] = df['current_year'].apply(clean_column)
```

```python
#cleaning the cgpa column

df['cgpa'] = df['cgpa'].str.strip()
```

```python
# plot of top 10 courses

plt.figure(figsize=(15,10))
sns.set_style('darkgrid')
sns.barplot(x = excel.course.value_counts().keys()[:10],y = excel.course.value_counts().head(10),edgecolor="black")
plt.title("Distribution of Top 10 First Class Course",fontweight="bold",fontsize="19",loc="center")
plt.show()
```

>There are more than 10 courses that have first classes. Specifically, in the BCS program, there are 14 students who achieved a first class. Similarly, in engineering, there are 9 first-class students, while psychology and BIT have 3 each.
>
>Focusing on these first four courses is beneficial for further analysis as they encompass a significant portion of our data.

```python
#The first two course

top4 = excel[(excel.course == 'engineering') | (excel.course =='bcs') | (excel.course =='psychology') | (excel.course =='bit')]
top4
```

```python
sns.set_style('darkgrid')
sns.countplot(x = top4.course,hue = excel.current_year,edgecolor="black")
plt.title("Course by Year of Study",fontweight="bold",fontsize="15",loc="center")
plt.show()
```

>Analyzing grades of engineering, psychology and computer science students shows something interesting: many get the highest marks (first-class distinctions) mostly in their first year. This pattern suggests that doing well early on plays a big role in how these students stand out.
